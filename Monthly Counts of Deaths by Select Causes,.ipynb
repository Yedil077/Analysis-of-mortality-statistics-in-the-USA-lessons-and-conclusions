```python
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

df = pd.read_csv('D:\DOWNLOADS\Cats\Monthly_Counts_of_Deaths_by_Select_Causes__2014-2019.csv')  # data processing, CSV file I/O (e.g. pd.read_csv)
df 
```

```python
plt.figure(figsize=(14, 8))
sns.lineplot(x='Year', y='All Cause', data=df, ci=None)
plt.title('Жылдар бойынша өлім-жітім динамикасы')
plt.xlabel('Жыл')
plt.ylabel('Өлім саны')
plt.show()
```

```python
plt.figure(figsize=(14, 8))
df_causes = df.drop(['Jurisdiction of Occurrence', 'Year', 'Month', 'All Cause', 'Natural Cause'], axis=1)
sns.barplot(data=df_causes, ci=None)
plt.title('Өлім себептерін бөлу')
plt.xlabel('Өлімнің себебі')
plt.ylabel('Өлім саны')
plt.xticks(rotation=90)
plt.show()
```

```python
plt.figure(figsize=(14, 8))
df_causes_monthly = df.groupby('Month').sum().reset_index()
sns.lineplot(x='Month', y='All Cause', data=df_causes_monthly, ci=None)
plt.title('Өлім себептерін айлар бойынша салыстыру')
plt.xlabel('Ай')
plt.ylabel('Өлім саны')
plt.show()
```

```python
plt.figure(figsize=(14, 8))
corr_matrix = df_causes.corr()
sns.heatmap(corr_matrix, annot=True, cmap='coolwarm', linewidths=.5)
plt.title('Өлім себептерінің өзара байланысы үшін жылу картасы')
plt.show()
```
